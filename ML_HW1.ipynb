```python
#a) finding likelihood function
def poisson_pmf(lambda_, k):
    return np.exp(-lambda_) * (lambda_**k) / np.math.factorial(k)

#using observations instead of k to make it obvious for myself what it means
def likelihood_function(lambda_, observations):
    likelihood = 1.00
    for obs in observations:
      likelihood *= poisson_pmf(lambda_, obs)
    return likelihood
```

```python
#testing my function to see if it works
observations = [2, 3, 4, 5, 3, 2]
likelihood_function(1,observations)
```

```python
def log_likelihood_function(lambda_,observations):
  log_likelihood = 0.0
  for obs in observations:
    log_likelihood += log(poisson_pmf(lambda_,obs))
  return log_likelihood

#testing my function
log_likelihood_function(1,observations)
```

```python
#for all poisson distributions, the maximum likelihood estimator lambda will be the mean of the observations!
#derivation for part c is on the next page
def MLE(observations):
    lambda_mle = sum(observations) / len(observations)
    return lambda_mle
```

```python
#a)
roll_1 = np.array([5, 3, 9, 3, 8, 4, 7])
roll_2 = np.array([5, 3, 9, 3, 8, 4, 7, 10])
freq_a = np.array([0,6,4,3,2,1,1,1,1,1,0]) #remember, 0 in index for convenience
freq_b = np.array([0,3,3,2,2,2,2,2,2,1,1])
freq_e = np.array([0,2,2,2,2,2,2,2,2,2,2])

cond_a = freq_a/np.sum(freq_a)
cond_b = freq_b/np.sum(freq_b)
cond_e = freq_e/np.sum(freq_e)
```

```python
import pandas as pd
import seaborn as sns

data = roll_1

p_a,p_b,p_e = [1/3], [1/3],[1/3]
for d in data:
    p_a.append(cond_a[d]*p_a[-1] / (cond_a[d]*p_a[-1] + cond_b[d]*p_b[-1]+cond_e[d]*p_e[-1]))
    p_b.append(cond_b[d]*p_b[-1] / (cond_a[d]*p_a[-1] + cond_b[d]*p_b[-1]+cond_e[d]*p_e[-1]))
    p_e.append(1-p_a[-1]-p_b[-1])
    # p_b.append(cond_dist_b[d]*p_b[-1] / (cond_dist_a[d]*p_a[-2] + cond_dist_b[d]*p_b[-1])) # note the -2

df = pd.DataFrame({'P(A|data)': p_a, 'P(B|data)': p_b,'P(E|data)': p_e})
df.loc[7]
df
```

c) dice A is eliminated from the pool of possibilities when calculating likelihood after rolling a 10 because we know that P(A|10)=0

```python
#P(B|E)
def simulation_e(simulations):
  results = np.array([])
  for i in np.arange(simulations):
    rolls = np.array([])
    p_b = 0.5
    p_e = 0.5
    while p_e <= 0.99:
      roll = np.random.choice(np.arange(11), p=cond_e)
      rolls = np.append(rolls, roll)
      latest_roll = int(rolls[-1])
      p_b = cond_b[latest_roll] * p_b / (cond_b[latest_roll] * p_b + cond_e[latest_roll] * p_e)
      p_e = 1 - p_b
    results = np.append(results,len(rolls))
  return results.mean()

def simulation_b(simulations):
  results = np.array([])
  for i in np.arange(simulations):
    rolls = np.array([])
    p_b = 0.5
    p_e = 0.5
    while p_b <= 0.99:
      roll = np.random.choice(np.arange(11), p=cond_b)
      rolls = np.append(rolls, roll)
      latest_roll = int(rolls[-1])
      p_b = cond_b[latest_roll] * p_b / (cond_b[latest_roll] * p_b + cond_e[latest_roll] * p_e)
      p_e = 1 - p_b
    results = np.append(results,len(rolls))
  return results.mean()
```

```python
simulation_e(1000)
```

```python
simulation_b(1000)
```

Using a simulation of 1000 rolls: it takes around 80 rolls for us to determine that the dice is E with 99% probability given that the real dice has the probability distribution of dice E, and it takes around 90 rolls for us to determine that the dice is B with 99% probability given that the real dice has the probability distribution of dice B. In order to test the true mean, we would need to create a confidence interval of sorts, but that is outside the bounds of this question.
